```python
from tensorflow.keras.preprocessing.text import one_hot
```

Sentences

```python
sentence=[
  'the glass of milk',
  'the glass of juice',
  'the cup of tea',
  'I am a good boy',
  'I am a good developer',
  'understand the meaning of words',
  'your videos are good',
]
```

```python
sentence
```

Define the Vacabulary Size

```python
voc_size=10000
```

One Hot Representation

```python
one_hot_rep=[one_hot(words,voc_size) for words in sentence]
one_hot_rep
```

Word Embedding Representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sentence_length=8
embedded_docs=pad_sequences(one_hot_rep,padding='pre',maxlen=sentence_length)
print(embedded_docs)
```

Feature Representation

```python
dimension=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dimension,input_shape=(sentence_length,)))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
